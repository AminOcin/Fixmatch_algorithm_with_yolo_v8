```python
import os
import shutil
import random

# Paths
train_images_path = '/content/drive/MyDrive/project/archive/train/images'
train_labels_path = '/content/drive/MyDrive/project/archive/train/labels'

# New folder paths
lab_train_images_path = '/content/drive/MyDrive/project/archive/lab_train/images'
lab_train_labels_path = '/content/drive/MyDrive/project/archive/lab_train/labels'

# Create new directories if they don't exist
os.makedirs(lab_train_images_path, exist_ok=True)
os.makedirs(lab_train_labels_path, exist_ok=True)


for filename in os.listdir(lab_train_images_path):
    file_path = os.path.join(lab_train_images_path, filename)
    if os.path.isfile(file_path):
        os.remove(file_path)


for filename in os.listdir(lab_train_labels_path):
    file_path = os.path.join(lab_train_labels_path, filename)
    if os.path.isfile(file_path):
        os.remove(file_path)

# List to hold valid image-label pairs
valid_pairs = []

# Check each label file and gather valid pairs
for filename in os.listdir(train_labels_path):
    if filename.endswith('.txt'):
        label_file_path = os.path.join(train_labels_path, filename)

        # Check if the label file is not empty
        if os.path.getsize(label_file_path) > 0:
            image_file_path = os.path.join(train_images_path, filename.replace('.txt', '.jpg'))  # Assuming images are .jpg

            # Check if the corresponding image exists
            if os.path.exists(image_file_path):
                valid_pairs.append((image_file_path, label_file_path))

# Calculate 20% of valid pairs to copy
num_to_copy = max(1, int(0.2 * len(valid_pairs)))  # Ensure at least one file is copied if there are valid pairs

# Shuffle the valid pairs
random.shuffle(valid_pairs)

# Copy the selected images and labels to the new folder
for i in range(num_to_copy):
    image_file, label_file = valid_pairs[i]
    shutil.copy(image_file, lab_train_images_path)
    shutil.copy(label_file, lab_train_labels_path)

# Summary
print(f"Copied {num_to_copy} valid image-label pairs to {lab_train_images_path} and {lab_train_labels_path}.")
```

plot of the classes in each folder train,test, valid


```python
import os
import yaml
import matplotlib.pyplot as plt
from collections import Counter

# Path to your dataset (adjust this if needed)
dataset_path = '/content/drive/MyDrive/project/archive'

# Load the YAML file to get class names
yaml_file = '/content/drive/MyDrive/project/archive/data.yaml'
with open(yaml_file, 'r') as f:
    data = yaml.safe_load(f)

# Extract class names and paths
class_names = data['names']
paths = {
    'train': data['train'].replace('..', dataset_path),
    'test': data['test'].replace('..', dataset_path),
    'valid': data['val'].replace('..', dataset_path)
}

# Define solid colors for train, test, and valid datasets
dataset_colors = {
    'train': 'blue',
    'test': 'green',
    'valid': 'red'
}

# Function to count class occurrences
def count_classes(labels_folder):
    class_counter = Counter()
    for label_file in os.listdir(labels_folder):
        if label_file.endswith('.txt'):
            with open(os.path.join(labels_folder, label_file), 'r') as f:
                for line in f:
                    class_id = int(line.split()[0])  # Class ID is the first number
                    class_counter[class_id] += 1
    return class_counter

# Initialize a Counter to hold counts for train, test, and valid sets
class_counts = {'train': Counter(), 'test': Counter(), 'valid': Counter()}

# Define paths for train, test, and valid labels
for dataset_type in ['train', 'test', 'valid']:
    labels_folder = os.path.join(paths[dataset_type].replace('/images', '/labels'))
    class_counts[dataset_type] = count_classes(labels_folder)

# Plot the class distribution with different solid colors for each dataset
def plot_class_distribution(class_counts, dataset_type):
    classes = [class_names[i] for i in range(len(class_names))]
    counts = [class_counts[dataset_type][i] for i in range(len(class_names))]
    color = dataset_colors[dataset_type]  # Use the color assigned to the dataset type

    plt.figure(figsize=(10, 6))
    plt.bar(classes, counts, color=color)  # Assign the solid color for the dataset type
    plt.title(f'Class Distribution in {dataset_type.capitalize()} Set')
    plt.xlabel('Classes')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()

# Plot for each dataset type
for dataset_type in ['train', 'test', 'valid']:
    plot_class_distribution(class_counts, dataset_type)
```

```python
from ultralytics import YOLO
import os
import cv2
import albumentations as A

# Load the pre-trained YOLOv8 model
model = YOLO('/content/drive/MyDrive/project/archive/runs/detect/train4/weights/best.pt')

# Directory paths
pseudo_label_dir = '/content/drive/MyDrive/project/archive/unlab_train/pseudo_level'
unlab_train_dir = '/content/drive/MyDrive/project/archive/unlab_train/images'

# Delete existing pseudo-label files
if os.path.exists(pseudo_label_dir):
    for file in os.listdir(pseudo_label_dir):
        file_path = os.path.join(pseudo_label_dir, file)
        if os.path.isfile(file_path):
            os.remove(file_path)
            print(f"Deleted: {file_path}")

# Set a lower confidence threshold
confidence_threshold = 0.7

# Define weak augmentation pipeline
weak_augmentation = A.Compose([
    A.HorizontalFlip(p=0.5),  # Randomly flip the image horizontally
    A.RandomBrightnessContrast(p=0.2, brightness_limit=0.5),  # Random brightness and contrast adjustment
    A.Resize(height=640, width=640)  # Resize the image
])

# Loop through unlabeled images and generate pseudo-labels
for img_file in os.listdir(unlab_train_dir):
    if img_file.endswith(('.jpg', '.jpeg', '.png')):
        img_path = os.path.join(unlab_train_dir, img_file)

        # Read the image using OpenCV
        image = cv2.imread(img_path)

        # Apply weak augmentation
        augmented = weak_augmentation(image=image)
        augmented_image = augmented['image']

        # Predict using the YOLOv8 model on the augmented image
        results = model.predict(augmented_image, conf=confidence_threshold)

        # Save predictions as pseudo-labels
        pseudo_label_path = os.path.join(pseudo_label_dir, img_file.replace('.jpg', '.txt'))
        with open(pseudo_label_path, 'w') as f:
            for box in results[0].boxes:
                cls = int(box.cls)  # class index
                x, y, w, h = box.xywh[0].tolist()  # box coordinates in YOLO format
                conf = box.conf[0].item()  # confidence score

                # Write to file if confidence is above the threshold
                if conf > confidence_threshold:
                    f.write(f"{cls} {x} {y} {w} {h}\n")

        print(f"Pseudo-labels saved for {img_file} at {pseudo_label_path}")
```

# Strong augment


```python
from ultralytics import YOLO
import os
import cv2
import albumentations as A

# Load the pre-trained YOLOv8 model
model = YOLO('/content/drive/MyDrive/project/archive/runs/detect/train4/weights/best.pt')

# Directory paths
unlab_train_dir = '/content/drive/MyDrive/project/archive/unlab_train'
image_dir = os.path.join(unlab_train_dir, 'images')
predictions_dir = os.path.join(unlab_train_dir, 'predictions')  # Specify where to save predictions

# Create predictions directory if it doesn't exist
os.makedirs(predictions_dir, exist_ok=True)

# Delete existing prediction files
for file in os.listdir(predictions_dir):
    if file.endswith('.txt'):
        os.remove(os.path.join(predictions_dir, file))

# Set a confidence threshold
confidence_threshold = 0.7

# Define strong augmentation pipeline
strong_augmentation = A.Compose([
    A.HorizontalFlip(p=0.5),  # Randomly flip the image horizontally
    A.RandomBrightnessContrast(p=0.2, brightness_limit=0.5),  # Random brightness and contrast adjustment
    A.Rotate(limit=30, p=0.5),  # Random rotation
    A.ShiftScaleRotate(shift_limit=0.1, scale_limit=0.1, rotate_limit=30, p=0.5),  # Random shifting, scaling, and rotating
    A.Resize(height=640, width=640)  # Resize the image
])

# Loop through unlabeled images and generate predictions
for img_file in os.listdir(image_dir):
    if img_file.endswith(('.jpg', '.jpeg', '.png')):
        img_path = os.path.join(image_dir, img_file)

        # Read the image using OpenCV
        image = cv2.imread(img_path)

        # Apply strong augmentation
        augmented = strong_augmentation(image=image)
        augmented_image = augmented['image']

        # Predict using the YOLOv8 model on the augmented image
        results = model.predict(augmented_image, conf=confidence_threshold)

        # Convert the generator to a list to access the first result
        results_list = list(results)

        # Save predictions as text files in the specified predictions directory
        prediction_path = os.path.join(predictions_dir, img_file.replace('.jpg', '.txt').replace('.jpeg', '.txt').replace('.png', '.txt'))
        with open(prediction_path, 'w') as f:
            for box in results_list[0].boxes:  # Access the first result
                cls = int(box.cls)  # class index
                x, y, w, h = box.xywh[0].tolist()  # box coordinates in YOLO format
                conf = box.conf[0].item()  # confidence score

                # Write to file if confidence is above the threshold
                if conf > confidence_threshold:
                    f.write(f"{cls} {x} {y} {w} {h}\n")

        print(f"Predictions saved for {img_file} at {prediction_path}")
```

```python
import os
import torch

pseudo_dir = '/content/drive/MyDrive/project/archive/unlab_train/pseudo_label'   # Directory containing pseudo labels (txt files)
pred_dir = '/content/drive/MyDrive/project/archive/unlab_train/predictions'       # Directory containing predictions (txt files)
```

```python
def read_label_file_to_tensor(file_path):
    # Read file and convert each line to a list of [class, x_center, y_center, width, height]
    with open(file_path, 'r') as f:
        labels = []
        for line in f.readlines():
            parts = line.strip().split()
            if len(parts) == 5:  # Ensure it's formatted as class, x, y, w, h
                labels.append([float(part) for part in parts])  # Convert all to float
        # Convert list to torch.Tensor
        if labels:
            return torch.tensor(labels)  # shape: [num_boxes, 5]
        else:
            return None  # In case file is empty or badly formatted
```

```python
pseudo_tensors = {}
pred_tensors = {}

# Iterate over pseudo-labels and predictions directories
for filename in os.listdir(pseudo_dir):
    if filename.endswith('.txt'):
        pseudo_file_path = os.path.join(pseudo_dir, filename)
        pred_file_path = os.path.join(pred_dir, filename)

        # Read pseudo-labels and predictions
        pseudo_tensor = read_label_file_to_tensor(pseudo_file_path)
        pred_tensor = read_label_file_to_tensor(pred_file_path)

        if pseudo_tensor is not None and pred_tensor is not None:
            pseudo_tensors[filename] = pseudo_tensor
            pred_tensors[filename] = pred_tensor
        else:
            print(f"Skipping {filename}, as one of the files is empty or invalid.")
```

```python
import torch.nn.functional as F

# Initialize variables to accumulate loss and count the number of files
total_consistency_loss = 0.0
num_files = 0

for filename in pseudo_tensors:
    pseudo_boxes = pseudo_tensors[filename][:, 1:]  # Extract x, y, w, h (ignore class)
    pred_boxes = pred_tensors[filename][:, 1:]  # Extract x, y, w, h (ignore class)

    # Check if shapes match and handle padding if necessary
    if pseudo_boxes.shape != pred_boxes.shape:
        max_size = max(pseudo_boxes.shape[0], pred_boxes.shape[0])

        # Pad pseudo_boxes if needed
        if pseudo_boxes.shape[0] < max_size:
            padding = torch.zeros(max_size - pseudo_boxes.shape[0], pseudo_boxes.shape[1])
            pseudo_boxes = torch.cat([pseudo_boxes, padding], dim=0)

        # Pad pred_boxes if needed
        if pred_boxes.shape[0] < max_size:
            padding = torch.zeros(max_size - pred_boxes.shape[0], pred_boxes.shape[1])
            pred_boxes = torch.cat([pred_boxes, padding], dim=0)

    # Compute consistency loss (L1 loss here) for the current file
    consistency_loss = F.l1_loss(pred_boxes, pseudo_boxes)
    print(f"Consistency Loss (BBox Loss) for {filename}: {consistency_loss.item()}")

    # Accumulate the total consistency loss
    total_consistency_loss += consistency_loss.item()
    num_files += 1

# Compute average consistency loss if there are any files processed
if num_files > 0:
    avg_consistency_loss = total_consistency_loss / num_files
    print(f"\nAverage Consistency Loss (BBox Loss): {avg_consistency_loss}")
else:
    print("No files processed.")
```

```python
##accuracy

import torch

def calculate_iou(box1, box2):
    """Calculate Intersection over Union (IoU) between two bounding boxes."""
    # Unpack boxes
    x1_min, y1_min, w1, h1 = box1
    x2_min, y2_min, w2, h2 = box2

    # Calculate coordinates of the intersection rectangle
    x1_inter = max(x1_min, x2_min)
    y1_inter = max(y1_min, y2_min)
    x2_inter = min(x1_min + w1, x2_min + w2)
    y2_inter = min(y1_min + h1, y2_min + h2)

    # Calculate area of intersection
    inter_area = max(0, x2_inter - x1_inter) * max(0, y2_inter - y1_inter)

    # Calculate area of both bounding boxes
    box1_area = w1 * h1
    box2_area = w2 * h2

    # Calculate IoU
    iou = inter_area / float(box1_area + box2_area - inter_area + 1e-6)  # Add epsilon to prevent division by zero
    return iou

# Initialize accuracy variables
total_true_positives = 0
total_predictions = 0

# IoU threshold for determining true positives
iou_threshold = 0.5

for filename in pseudo_tensors:
    pseudo_boxes = pseudo_tensors[filename][:, 1:]  # Extract x, y, w, h (ignore class)
    pred_boxes = pred_tensors[filename][:, 1:]  # Extract x, y, w, h (ignore class)

    # Check if shapes match and handle padding if necessary
    if pseudo_boxes.shape != pred_boxes.shape:
        max_size = max(pseudo_boxes.shape[0], pred_boxes.shape[0])

        # Pad pseudo_boxes if needed
        if pseudo_boxes.shape[0] < max_size:
            padding = torch.zeros(max_size - pseudo_boxes.shape[0], pseudo_boxes.shape[1])
            pseudo_boxes = torch.cat([pseudo_boxes, padding], dim=0)

        # Pad pred_boxes if needed
        if pred_boxes.shape[0] < max_size:
            padding = torch.zeros(max_size - pred_boxes.shape[0], pred_boxes.shape[1])
            pred_boxes = torch.cat([pred_boxes, padding], dim=0)

    # Count predictions
    total_predictions += pred_boxes.shape[0]

    for pseudo_box, pred_box in zip(pseudo_boxes, pred_boxes):
        # Compute IoU for each box pair
        iou = calculate_iou(pseudo_box, pred_box)

        # Check if IoU exceeds the threshold
        if iou >= iou_threshold:
            total_true_positives += 1

# Calculate accuracy
if total_predictions > 0:
    accuracy = total_true_positives / total_predictions
    print(f"Accuracy: {accuracy * 100:.2f}%")
else:
    print("No predictions to evaluate.")
```
